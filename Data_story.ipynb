***We collected the dataset, containing the number of students enrolled in Class-I to Class-V in various states of India.Following are the plots representing this data and their corresponding interpretation.***

## Scatter Plot:

```python
plt.figure(figsize=(10,5))
plt.grid()
plt.scatter(data.loc[:,"Enrolment in Class I - Boys"],data.loc[:,"Enrolment in Class I - Girls"],marker='*',color="green",label="Class-I")
plt.scatter(data.loc[:,"Enrolment in Class II - Boys"],data.loc[:,"Enrolment in Class II - Girls"],marker='+',color="blue",label="Class-II")
plt.scatter(data.loc[:,"Enrolment in Class III - Boys"],data.loc[:,"Enrolment in Class III - Girls"],color="yellow",label="Class-III")
plt.scatter(data.loc[:,"Enrolment in Class IV - Boys"],data.loc[:,"Enrolment in Class IV - Girls"],color="orange",label="Class-IV")
plt.scatter(data.loc[:,"Enrolment in Class V - Boys"],data.loc[:,"Enrolment in Class V - Girls"],marker='x',color="red",label="Class-V")
plt.legend()
plt.xlabel("Number of Boys Enrolled")
plt.ylabel("Number of Girls Enrolled")
```

### **Interpretation**:
*We observe that over all the states, number of students enrolled is the highest in Class-I,shown by the green star at the top-right.
And this number decreases as we move along the classes.At class-II and class-III numbers are almost the same, but falls down significantly in Class-IV then to Class-V.*

---
*Since the plot is nearly linear with a slope of 45-degrees, this tells us that almost in all Indian states,gender-distribution is uniform in children enrolled in the primary education.*
